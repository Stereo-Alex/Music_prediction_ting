## 2. Cluster recomender system

```python
## giving a song recomendation in base of the song the user wrote in the first function 

#first funtion just fetches the song features and removes clutter 

def conection_api_user_song(user_input_song):
    
    user_input_song_2 = sp.search(user_input_song)
    features = sp.audio_features(user_input_song_2["tracks"]["items"][0]["uri"])
    temp_df = pd.DataFrame(features)
    temp_df = temp_df.drop([ "uri", "analysis_url", "id", "track_href", "type", "time_signature", "duration_ms"], axis=1)
    temp_df_f = temp_df.loc[:, ~temp_df.columns.str.contains('^Unnamed')]
   
    return temp_df_f

```

```python
# we import our pickeled variables 

import pickle
scaler_final = pickle.load(open("scaler_final.pkl", "rb"))

#one is our scaler and the other the actual model 

Model_predictor = pickle.load(open("KMeans_final.pkl", "rb"))
```

```python
# we must now transform using the scaler we imported the song we obtained from the user 
#where tester_song is the output of conection_api_user_song(user_input_song)


#takes user input and returns an array to be used later 

def song_to_array(user_input):
    pre_transform = conection_api_user_song(user_input)
    post_scale = scaler_final.transform(pre_transform)
    return post_scale

    
    
#song_to_array()
```

```python
# we now import our data frame to be able to see where our request falls in our clusters 

data_with_clusters = pd.read_csv("final_data_clusters.csv")


data_with_clusters.head(50)
```

```python
#we now input our song prediction into a function that will return a recomended song from the same cluster  


def song_recomender(song_prediction):
    filtered_df = data_with_clusters["clusters"] == song_prediction[0]
    data_with_clusters_filtered = data_with_clusters[filtered_df]
    recomended_song_df = data_with_clusters_filtered.sample()
    final_output = print(f"Your recomended song is {recomended_song_df.iloc[0].at['Songs']}"
                         f" by {recomended_song_df.iloc[0].at['Artists']} "
                         f"\nfind it  by pasting this link in your browser {recomended_song_df.iloc[0].at['uri']}")
    return final_output

```
